### 4.4. Коэффициент удержания.

Т.к. это агрегатор по продаже билетов и вряд-ли много людей пользуется им ежедневно, да и еженедельно тоже. Лучший вариант будет построить когорты периодом в 1 месяц.

```python
#  добавим дату первой активности клиента
visits['start_ts_date'] = pd.to_datetime(visits['start_ts_date'])
first_activity_visits = visits.groupby(['uid'])['start_ts'].min()
first_activity_visits.name = 'first_activity_date'
visits = visits.join(first_activity_visits, on='uid')

#  выделим из столбцов first_activity_date и start_ts_date месяцы.
visits['first_activity_date'] = pd.to_datetime(visits['first_activity_date'])
visits['first_activity_month'] = visits['first_activity_date'].astype('datetime64[M]')
visits['activity_month'] = visits['start_ts_date'].astype('datetime64[M]')

#  для каждой строки датафрейма рассчитаем lifetime пользователя в рамках когорты
visits['cohort_lifetime'] = visits['activity_month'] - visits['first_activity_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')

#  груперуем данные по кагортам
cohorts = (visits.groupby(['first_activity_month','cohort_lifetime'])
           .agg({'uid':'nunique'}).reset_index())

#  получаем число пользователей, изначально бывших в когорте, 
#  и на него делим число пользователей в каждый следующующий месяц
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(initial_users_count,on='first_activity_month')
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']

#  строим сводную таблицу отсеим первый месяц жизни, т.к. там все значения будут 100%
retention_pivot = (cohorts.query('cohort_lifetime != 0')
                   .pivot_table(
                       index='first_activity_month',
                       columns='cohort_lifetime',
                       values='retention',
                       aggfunc='sum'))
retention_pivot.index = retention_pivot.index.strftime("%m-%Y")

#  создаём тепловую карту
sns.set(style='white')
plt.figure(figsize=(16, 8))
plt.title('Коэффициент удержания когорт по месяцу')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', 
            vmin=0, vmax=0.1, 
            linewidths=1, linecolor='gray')
plt.show()
```

Найдём средний Retention Rate (Коэффициент удержания) на второй месяц "жизни" когорты:

```python
print('Retention Rate = {:.1%}'.format(retention_pivot[1].mean()))
```

### Вывод.

- На второй месяц остаётся в среднем 6.5% активных пользователей.
- Начиная с когорты **09-2017** коэффициент удержания с течением времени только падает.
- Начиная с когорты **12-2017** коэффициент удержания со второго месяца очень снизился.

В целом Retention Rate становится меньше, возможно это связанно с сезонностью, т.к. мы наблюдаем спад после декабря, либо была применена новая рекламная кампания, которая не показывает себя успешно. Возможно стоит продумать иные способы привлечения бывших пользователей.

## 5. Метрики электронной коммерции.

### 5.1. Первая покупка.

Исследуем, сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки.

```python
#  найдём дату первой покупки
first_orders = orders.groupby('uid')['buy_ts'].min()
first_orders.name = 'first_orders_date'
orders = orders.join(first_orders, on='uid')

#  объеденим таблицу с первым посещением сайта и первой покупкой
#  дата первого посещения сайта у нас уже есть в переменной first_activity_visits
average_first_orders = pd.merge(first_activity_visits, first_orders, on='uid').reset_index()

#  проверим как подгрузилось
average_first_orders.head()
```

Проверим есть ли у нас покупки совершённые раньше чем первый визит на сайт.

```python
average_first_orders.query('first_orders_date < first_activity_date').count()
```

Посчитаем среднее время покупки.

```python
average_first_orders['average_first_orders'] = (average_first_orders['first_orders_date'] 
                                                - average_first_orders['first_activity_date'])

#  выведем для сравнения среднее и медиану времени первой покупки
print('Среднее время с момента первого посещения сайта до покупки:', 
      average_first_orders['average_first_orders'].mean())
print()
print('Медиана с момента первого посещения сайта до покупки:',
     average_first_orders['average_first_orders'].median())
```

Посмотрим на распределение среднего времени до первой покупки (в днях). Т.к. у нас половине пользователей хватает 16 минут, а в среднем пользователи. тратят почти 17 дней, рассмотрим диапазон в первый меясц (30 дней).

```python
# переводим в кол-во дней
average_first_orders['average_first_orders'] = average_first_orders['average_first_orders'] / np.timedelta64(1, 'D')
average_first_orders['average_first_orders'] = average_first_orders['average_first_orders'].round().astype('int')

# построим гистограмму
average_first_orders['average_first_orders'].hist(bins=100, range=(0,30) ,figsize=(14,7))
plt.title('Распределение от первого посещения до первой покупки')
plt.xlabel('Время до совершения покупки')
plt.ylabel('Кол-во покупок')
plt.show()
```

### 5.2. Сколько покупок за период?

### 5.4. Ценность клиента.

Посмотрим, как меняется LTV (сумма денег, которую один клиент приносит в среднем со всех своих покупок) на покупателя со временем.

Из предоставленных данных мы знаем что маржинальность - 100%.

```python
#  добавим дату совершения покупки
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

#  вычислим, сколько людей совершили покупку впервые в каждом месяце
cohort_sizes = orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

#  маржинальность магазина - 100%
margin_rate = 1

#  считаем LTV
cohorts_orders = orders.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()
report = pd.merge(cohort_sizes, cohorts_orders, on='first_order_month')
report['gp'] =  report['revenue'] * margin_rate
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report['ltv'] = report['gp'] / report['n_buyers']

#  считаем среднее LTV по кагортам
result = report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean')
result.index = result.index.strftime("%m-%Y")

#  создаём тепловую карту
sns.set(style='white')
plt.figure(figsize=(16, 8))
plt.title('LTV (ценность клиента)')
sns.heatmap(result.head(12).cumsum(axis = 1), annot=True, linewidths=1, linecolor='gray')
plt.show()
```

```python
# считаем LTV средней когорты за 6 месяцев
m6_cum_ltv = result.cumsum(axis=1).mean(axis=0)[5]

print('Средний LTV за 6 месяцев после первой покупки:', m6_cum_ltv.round(2), 'y.e.')
```

### Вывод.

За пол года в среднем покупатели приносят 7.97 y.e.. 

Судя по карте когорты приносят основную прибыль в первый месяц, далее малая часть пользователей возвращается ежемесячно делать заказы на сервисе, но бывают "всплески" (возможно какие-нибудь акции, либо премьеры).

## Общий вывод.

На входе клиенты приносят основную часть прибыли. А так же в период с сентября по декабрь. 

Но можно отметить тех пользователей, что пришли с **6-2017 по 9-2017**, т.к. со временем жизни LTV этих когорт увеличился больше остальных.

### 6.2. Привлечение одного покупателя.

Посчитаем стоимость привлечения клиента по каждому источнику трафика и среднее на одного покупателя.

```python
#   найдём для каждого пользователя главный(первый) рекламный источник
uid_info = (visits.sort_values('start_ts', ascending = True)
            .groupby(['uid']).agg({'source_id' : 'first'}).reset_index())
```

```python
#  объединим таблицу с информацией о заказх
orders = pd.merge(orders, uid_info, on='uid')
```

```python
#  найдём количество людей в когортах
source_cohort_sizes = (orders.groupby(['source_id', 'first_order_month'])
                       .agg({'uid': 'nunique'}).reset_index())
source_cohort_sizes.columns = ['source_id', 'first_order_month', 'n_buyers']
```

```python
# считаем сумму заказов
cohorts_source = (orders.groupby(['source_id', 'first_order_month', 'order_month'])
                  .agg({'revenue': 'sum'}).reset_index())
```

```python
#  объединяем получившиеся таблицы
final_report = source_cohort_sizes.merge(cohorts_source, 
                                         left_on = ['source_id', 'first_order_month'], 
                                         right_on = ['source_id', 'first_order_month'])
```

```python
#  считаем LTV
final_report['gp'] =  final_report['revenue'] * margin_rate
final_report['age'] = (final_report['order_month'] - final_report['first_order_month']) / np.timedelta64(1, 'M')
final_report['age'] = final_report['age'].round().astype('int')
final_report['ltv'] = final_report['gp'] / final_report['n_buyers']
```

```python
#  считаем расходны по источнику и месяцам
monthly_costs = costs.groupby(['source_id', 'dt_month']).agg({'costs': 'sum'}).reset_index()
```

```python
#  посчитаем CAC
report_new = pd.merge(final_report, monthly_costs, 
                      left_on=['source_id', 'first_order_month'], 
                      right_on=['source_id', 'dt_month'])
report_new['cac'] = report_new['costs'] / report_new['n_buyers']
```

Посмотрим как меняется CAC для каждого источника трафика со временем.

```python
report_new.pivot_table(index = 'first_order_month', 
                       columns = 'source_id', 
                       values = 'cac', 
                       aggfunc = 'mean').plot(grid=True, figsize=(16,8))
plt.title('Динамика CAC для каждого источника трафика')
plt.xlabel('Периоды')
plt.ylabel('CAC')
plt.show()
```

```python
(report_new.pivot_table(index = 'source_id', values = 'cac', aggfunc = 'mean')
 .round(2)
 .sort_values(by='cac', ascending=False)
 .reset_index())
```

```python
print('Средний CAC на одного покупателя:', report_new['cac'].mean().round(2), 'y.e.')
```

### Вывод.

Источники 3 и 2, по стоимости на клиента, обходятся выше чем ценность клиента (7.97 y.e.).

Однако средний CAC на одного покупателя 8.5 y.e. но это тоже больше LTV клиента.
